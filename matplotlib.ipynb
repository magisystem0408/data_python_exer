```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot([1,2,3,4])
plt.ylabel('you are mamushi?')
plt.show()
```

```python
plt.plot([1,2,3,4],[1,4,9,16])
plt.show()
```

```python
# 2軸目で形状を変えられる
plt.plot([1,2,3,4],[1,4,9,16],'ro')
# どこまで描画するか
plt.axis([0,6,0,10])
plt.show()
```

```python
t =np.arange(0,5,0.2)
plt.plot(t,t,'r--',t,t**2,'bs',t,t**3,'g^')
plt.show()
```

```python
x=np.random.rand(50)
y=np.random.rand(50)

colors =np.random.rand(50)
# 大きさ指定
area =np.random.rand(50)*100

plt.scatter(x,y,s=area,c =colors,alpha=0.5)
plt.show()
```

```python
# 棒グラフの描画
objects =('a','b','c','d','e','f')
y_pos =np.arange(len(objects))
value=[1,2,3,4,5,6]

# データ突っ込む
plt.bar(y_pos,value,alpha=0.5)

# どうやって描画するか
plt.xticks(y_pos,objects)

plt.ylabel('Usage')
plt.title('title')
plt.show()
```

```python
# 円グラフ

labels=('python','C++','RUby','Java')
size=[10,20,30,40]
colors=['red','green','yellow','blue']


plt.pie(size,labels=labels,
        colors=colors, 
        
        #%%はワイルドカードなのでスキップしている
        
        autopct='%1.1f%%',shadow=True)

# 軸を綺麗にしてあげる
plt.axis('equal')
plt.show()
```

```python
y =[2,4,6,8,10]
y2=[10,11,12,13,14]
x=np.arange(5)

plt.plot(x,y,label='Y')
plt.plot(x,y2,label='Y2')

# 説明文を付け加えるには、legen
plt.legend()
plt.show()
```
