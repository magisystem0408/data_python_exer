```python
import numpy as np
import pandas as pd
from pandas import DataFrame
```

```python
df =DataFrame({'key':['X','Y','Z','X','X','P'],'data_set_1':np.arange(6)})
df
```

```python
df2 =DataFrame({'key':['X','Y','Z'],'data_set_2':[1,2,3]})
df2
```

```python
# データマージ
# keyを使用してマージをする
pd.merge(df,df2,on='key')
```

```python
# howでどちらのdfを基準にしてmergeするかを指定する
pd.merge(df,df2,on='key',how='left')

# howの引数は
# left：第一引数
# right：第二引数
# outer：両方
```

```python
df3 =DataFrame({'key':['X','X','Y','Z','Z','Z'],'data_set_3':range(6)})
df4 =DataFrame({'key':['X','X','Z','Z','Z'],'data_set_4':range(5)})
```

```python
pd.merge(df3,df4)
```
